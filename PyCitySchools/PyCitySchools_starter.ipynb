```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data.head(2)
```

```python
total_schools = school_data["School ID"].count()
total_students = school_data["size"].sum()
total_budget = school_data["budget"].sum()
```

```python
school_data_complete.head(2)
```

```python
average_math = school_data_complete["math_score"].mean()
average_reading = school_data_complete["reading_score"].mean()
average_overall = (average_math + average_reading) / 2
```

```python
math_70 = school_data_complete.loc[:, "math_score"] >= 70
math_70_students = math_70.sum()
reading_70 = school_data_complete.loc[:, "reading_score"] >= 70
reading_70_students = reading_70.sum()

percentage_math_70 = (math_70_students / total_students)*100
percentage_reading_70 = (reading_70_students / total_students)*100
```

```python
reading_70_students
```

```python
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math],
    "Average Reading Score": [average_reading],
    "% Passing Math": [percentage_math_70],
    "% Passing Reading": [percentage_reading_70],
    "& Overall Passing Rate": [average_overall]
})
```

```python
district_summary
```

```python
index = school_data.set_index(["school_name"])["type"]
index.head()
```

```python
total_students_school = school_data_complete["school_name"].value_counts()
total_students_school.head()
```

```python
total_budget_school = school_data_complete.groupby("school_name").mean()["budget"]
total_budget_school.head()
```

```python
per_student_budget = total_budget_school / total_students_school
per_student_budget.head()
```

```python
average_math_bySchool = school_data_complete.groupby("school_name").mean()["math_score"]
print(average_math_bySchool)

average_reading_bySchool = school_data_complete.groupby("school_name").mean()["reading_score"]
print(average_reading_bySchool)
```

```python
math_percentage_70 = school_data_complete[(school_data_complete["math_score"] >= 70)]
percentage_math_bySchool = math_percentage_70.groupby("school_name").count()["student_name"]
reading_percentage_70 = school_data_complete[(school_data_complete["reading_score"] >= 70)]
percentage_reading_bySchool = reading_percentage_70.groupby("school_name").count()["student_name"]
```

```python
percent_in_percent_math = (percentage_math_bySchool / total_students_school)*100
percent_in_percent_reading = (percentage_reading_bySchool / total_students_school)*100
```

```python
overall_rate = (percent_in_percent_math + percent_in_percent_reading) / 2
```

```python
BySchool_summary = pd.DataFrame({
   "School Type": index,
   "Total Students": total_students_school,
   "Total School Budget": total_budget_school,
   "Per Student Budget": per_student_budget,
   "Average Math Score": average_math_bySchool,
   "Average Reading Score": average_reading_bySchool,
   "% Passing Math": percent_in_percent_math,
   "% Passing Reading": percent_in_percent_reading,
   "Overall Passing Rate": overall_rate
})
```
